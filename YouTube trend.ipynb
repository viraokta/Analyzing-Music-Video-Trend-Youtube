```python
from pytrends.request import TrendReq
#pandas handel the dataframe
import pandas as pd
#create the charts
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
kw_list = ["mv", "music video"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='ID', gprop='youtube')
```

```python
df_queries = pytrends.related_queries()
```

```python
top_music_video = df_queries.get("mv").get("top")
top_mv = df_queries.get("music video").get("top")
df_top = pd.concat([top_music_video, top_mv] )
df_top.sort_values(['value'], ascending = False).head(5).reset_index(drop = True)
```

```python
rising_music_video = df_queries.get("mv").get("rising")
rising_mv = df_queries.get("music video").get("rising")
df_rising = pd.concat([rising_music_video, rising_mv] )
df_rising.sort_values(['value'], ascending = False).head(5)
```

```python
kw_list = ["blackpink mv", "music 2018", "mv nct", "nct", "mv fake love"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='youtube')
df_interest = pytrends.interest_over_time().drop(columns='isPartial')
```

```python
kw_list = ["nct mv"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='youtube')
df_interest_region = pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=False)
df = df_interest_region.sort_values(by='nct mv', ascending=False).head(10).reset_index()
```

```python
fig = plt.figure(figsize=(15,5))
plt.bar( df['geoName'],
         df['nct mv'],
         color = ['#fa9d96','#fa928a','#ff828a','#ff728a','#ff628a','#ff528a','#ff428a',
                  '#ff328a','#ff228a','#ff028a'])
plt.xticks(rotation=45,ha='right')
plt.title('Countries Which Most Interested in NCT Music Video',y=1.1,fontsize=20)
plt.xlabel('Country')
plt.ylabel('Value')
```
